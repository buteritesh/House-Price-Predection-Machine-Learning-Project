```python
#Here we have diffrents kind of parameters in data sets means they are not correralted to each other so we go to feature scaling 
#The data.
```

```python
sc=StandardScaler()
sc.fit(X_train)#Only training data need to fit 
X_train=sc.transform(X_train)#Both training and test data need to transform
X_test=sc.transform(X_test)  
```

```python
X_train 
```

```python
reg=SVR(kernel='rbf')
```

```python
reg.fit(X_train,y_train)
```

```python
reg.score(X_train,y_train)
```

```python
reg.score(X_test,y_test)  
```

```python
#Accuracy is low so change your kerenel because it is linear data sets
```

```python
reg_l=SVR(kernel='linear')
```

```python
reg_l.fit(X_train,y_train)
```

```python
reg_l.score(X_test,y_test)  
```

```python
y_pred=reg_l.predict(X_test)
```

```python
result_df=pd.DataFrame({'Actual':y_test,'predicted':y_pred})
```
